```python
X = DatosSelec[['year']]
y = DatosSelec['selling_price']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)

beta_0 = modelo.intercept_
beta_1 = modelo.coef_

print(f"Intercepto (β0): {beta_0}")
print(f"Coeficiente (Modelo, β1): {beta_1}")

plt.scatter(DatosSelec['year'], y, color='blue', label='Datos Reales')
plt.plot(DatosSelec['year'], modelo.predict(X), color= 'red', label='Línea de Regresión')
plt.title('Regresión Lineal: Precio vs Modelo Vehiculo Maruti Swift Dzire VDI')
plt.xlabel('Modelo')
plt.ylabel('Precio (USD)')
plt.legend()
plt.show()
```

```python
nuevo_X = np.array([[2020], [2022]])
predicciones = modelo.predict(nuevo_X)

print("\nPredicciones:")
for i, pred in enumerate(predicciones):
    print(f"Predicción {i+1} para (Modelo: {nuevo_X[i][0]}): {pred:.2f} USD")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = DatosSelec[['year']]
y = DatosSelec['selling_price']
modelo = LinearRegression()
modelo.fit(X, y)
predicciones = modelo.predict(X)

r2 = r2_score(y, predicciones)
mae = mean_absolute_error(y, predicciones)
mse = mean_squared_error(y, predicciones)
rmse = np.sqrt(mse)

print(f'R2 (Coeficiente de determinación): {r2}')
print(f'Error Absoluto Medio (MAE): {mae}')
```
